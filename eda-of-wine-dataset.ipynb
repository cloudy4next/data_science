## Most of the wines in this dataset seem to have quality scores of 5 or 6 .


#### we add a new feature called "Reviews" which divides the wines into "Good" or "Bad" giving them a value of "0" if wine quality is <= 6 or "1" if wine quality is >6.

```python
def review_column(df):
    reviews=[]
    for i in df["quality"]:
        if i <= 6:
            reviews.append(0)
        else:
            reviews.append(1)
    return reviews
df["Reviews"] = review_column(df)
```

## Correlation Plot :

```python
def correaltion(data):
    calculate_corr = data.corr()
    new_corr = calculate_corr.transpose().loc[:, ["quality"]].sort_values(by="quality",ascending=False)
    plt.subplots(figsize=(10,10))
    sns.heatmap(calculate_corr,annot=True)
    return new_corr

correaltion(df)
```

# Part-II : Machine Learning

```python
df= df.drop(['quality'], axis='columns')

y=df['Reviews']
x=df.drop(['Reviews'], axis='columns')

def logitmodel_one(x,y):    
    logitmodel=sm.Logit(y,x)
    result=logitmodel.fit()
    print(result.summary2())
    
def logitmodel_two(x,y): 
    X= sm.add_constant(x)
    logitmodel2=sm.Logit(y,X)
    result2=logitmodel2.fit()
    print(result2.summary2() )
```

```python
logitmodel_one(x,y)
```

```python
logitmodel_two(x,y)
```
